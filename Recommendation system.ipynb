### Collaborative Filtering-UBCF(cosine similarity score)

## Model Building

```python
ubcf=pd.pivot_table(data=book_data,values='Book.Rating',index='User_ID',columns='Book.Title').fillna(0)
ubcf.index=book_data.User_ID.unique()
ubcf
```

```python
cosine_distances=1-pairwise_distances(X=ubcf,metric='cosine')
cosine_df=pd.DataFrame(data=cosine_distances)
cosine_df.index=book_data.User_ID.unique()
cosine_df.columns=book_data.User_ID.unique()
cosine_df

```

```python
first_500_users=cosine_df.iloc[:500,:500]
first_500_users
```

```python
def get_recommendation(customer_id):
    tem=list(cosine_df.sort_values([customer_id],ascending=False).head(3).index)
    book_list=list(book_data[book_data['User_ID'].isin(tem)]['Book.Title'])
    recommended_books=set(book_list)-set(book_data[book_data['User_ID']==customer_id]['Book.Title'])
    return recommended_books
```
