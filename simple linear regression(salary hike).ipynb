### Salary_hike -> Build a prediction model for Salary_hike

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
salary_data=pd.read_csv("Salary_Data.csv")
salary_data
```

```python
sns.lmplot(x='YearsExperience',y='Salary',data=salary_data)
```

```python
model1=smf.ols(formula='YearsExperience~Salary',data=salary_data).fit()
```

```python
model1.summary()
```

```python
sns.distplot(salary_data["YearsExperience"])
plt.show()
```

```python
sns.distplot(salary_data["Salary"])
plt.show()
```

```python
model1.rsquared
```

# SQRT TRANSFORMATION

```python
salary_data['salary_sqrt']=np.sqrt(salary_data['Salary']).round(0)
salary_data
```

```python
sns.lmplot(x='YearsExperience',y='salary_sqrt',data=salary_data)
plt.show()
```

```python
sqrt_model=smf.ols('YearsExperience~salary_sqrt',data=salary_data).fit()
sqrt_model
```

```python
sqrt_model.summary()
```

```python
sqrt_model.rsquared
```

# LOG TRANSFORMATION

```python
salary_data['log_salary']=np.log(salary_data['Salary'])
salary_data
```

```python
sns.lmplot(x='YearsExperience',y='log_salary',data=salary_data)
plt.show()
```

```python
log_model=smf.ols(formula='YearsExperience~log_salary',data=salary_data).fit()
```

```python
log_model.summary()
```

```python
log_model.rsquared
```

## Among three models first one have highest  rsquare value so we choose first one
